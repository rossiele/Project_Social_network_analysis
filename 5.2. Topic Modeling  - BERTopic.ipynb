```python
from bertopic import BERTopic
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
import nltk
```

```python
df = pd.read_csv("data/df_coms_lou_0.csv")
del df['Unnamed: 0']
df.shape
```

```python
#filtro il df selezionando i tweet che hanno un minimo di interazione (in termini di likes e rt)
df_filter = df.loc[(df["likes_count"]>=1) & (df["retweets_count"]>= 1)]
df = df_filter
df.shape
```

```python
df1 = pd.read_csv("data/df_coms_lou_1.csv")
del df1['Unnamed: 0']
```

```python
df2 = pd.read_csv("data/df_coms_lou_2.csv")
del df2['Unnamed: 0']
```

```python
df3 = pd.read_csv("data/df_coms_lou_3.csv")
del df3['Unnamed: 0']
```

```python
df = df.drop_duplicates()
```

```python
df1 = df1.drop_duplicates()
```

```python
df2 = df2.drop_duplicates()
```

```python
df3 = df3.drop_duplicates()
```

```python
tweets_df = df.tweet.astype(str).to_list()
timestamp_df = df.date.to_list()
```

```python
tweets_df1 = df1.tweet.astype(str).to_list()
timestamp_df1 = df1.date.to_list()
```

```python
tweets_df2 = df2.tweet.astype(str).to_list()
timestamp_df2 = df2.date.to_list()
```

```python
tweets_df3 = df3.tweet.astype(str).to_list()
timestamp_df3 = df3.date.to_list()
```

```python
topic_model = BERTopic(language="english", nr_topics = 20) #20 argomenti pi첫 in voga
topics, probs = topic_model.fit_transform(tweets_df)
```

```python
topic_model1 = BERTopic(language="english", nr_topics = 20) #20 argomenti pi첫 in voga
topics1, probs1 = topic_model1.fit_transform(tweets_df1)
```

```python
topic_model2 = BERTopic(language="english", nr_topics = 20) #20 argomenti pi첫 in voga
topics2, probs2 = topic_model2.fit_transform(tweets_df2)
```

```python
topic_model3 = BERTopic(language="english", nr_topics = 20) #20 argomenti pi첫 in voga
topics3, probs3 = topic_model3.fit_transform(tweets_df3)
```
